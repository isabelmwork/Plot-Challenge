### Analysis

There were 248 mice that we analyzed. 

The Ramicane drug regimen had the smallest mean, median, variance, and standard error tumor volume. On the other hand, the Ketapril drug regimen had the largest mean, median, variance, and standard error tumor volume.

Capomulin and Ramicane had the largest number of observed mouse timepoints. All of the other drug regimens except Propriva had a smaller number of observed mouse timepoints, but stayed relatively close to eachother in number. 

It's important to note that percentages of female mice and male mice are almost equal. So, both genders are most likely represented well in the study.

Infubinol was the only drug regimen to have an outlier in final tumor volume. Between Capomulin, Infubinol, Ramicane, and Ceftamin, Capomulin and Ramicane had the lowest median. They also had observed final tumor volumes that were noticably smaller than that of Infubinol and Ceftamin. 

There is a rather strong correlation between average tumor volume and weight of mice on the Capomulin drug regimen. 

```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

## Prepare Data

```python
#Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

#Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#Combine the data into a single DataFrame
mouse_study = pd.merge(mouse_metadata, study_results, how = 'outer', on = 'Mouse ID')

#Display the data table for preview
mouse_study.head()
```

```python
mouse_study = mouse_study.rename(columns = {
    "Mouse ID": "Mouse_ID",
    "Drug Regimen": "Drug_Regimen",
    "Weight (g)": "Weight_(g)",
    "Tumor Volume (mm3)": "Tumor_Volume_(mm3)",
    "Metastatic Sites": "Metastatic_Sites"
})
```

```python
#Checking the number of mice.
unique_IDs = len(mouse_study["Mouse_ID"].unique())
unique_IDs
```

```python
#Our data should be uniquely identified by Mouse ID and Timepoint
#Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = mouse_study[mouse_study.duplicated(["Mouse_ID", "Timepoint"])]
duplicate_mouse = duplicates["Mouse_ID"].unique()[0]

#https://www.geeksforgeeks.org/find-duplicate-rows-in-a-dataframe-based-on-all-or-selected-columns/
```

```python
#Optional: Get all the data for the duplicate mouse ID. 
last_duplicate = len(duplicates.index) - 1
mouse_study.loc[(duplicates.index[0] - 1):duplicates.index[last_duplicate]]
```

```python
#Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study = mouse_study.loc[mouse_study["Mouse_ID"] != "g989"]

#Checking the number of mice in the clean DataFrame.
len(mouse_study["Mouse_ID"].unique())
```

## Summary Statistics

```python
# Use groupby and summary statistical methods to calculate the 
#following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_mouse_study = mouse_study.set_index("Drug_Regimen")
drug_mouse_study = drug_mouse_study.groupby("Drug_Regimen")
```

```python
mean_tumor_vol = drug_mouse_study["Tumor_Volume_(mm3)"].mean()
median_tumor_vol = drug_mouse_study["Tumor_Volume_(mm3)"].median()
variance_tumor_vol = drug_mouse_study["Tumor_Volume_(mm3)"].var()
std_tumor_vol = drug_mouse_study["Tumor_Volume_(mm3)"].std()
sem_tumor_vol = drug_mouse_study["Tumor_Volume_(mm3)"].sem()

#Assemble the resulting series into a single summary DataFrame.
summary_mouse_study = pd.DataFrame({
    "Mean_Tumor_Volume": mean_tumor_vol,
    "Median_Tumor_Volume": median_tumor_vol,
    "Tumor_Volume_Variance": variance_tumor_vol,
    "Tumor_Volume_Std._Dev.": std_tumor_vol,
    "Tumor_Volume_Std._Err.": sem_tumor_vol
})
summary_mouse_study
```

## Quartiles, Outliers and Boxplots

```python
#Calculate the final tumor volume of each mouse across four of 
#the treatment regimens:  
#Capomulin, Ramicane, Infubinol, and Ceftamin
best_drug_mouse = mouse_study.loc[(mouse_study["Drug_Regimen"] == "Capomulin") | (mouse_study["Drug_Regimen"] == "Ramicane") | (mouse_study["Drug_Regimen"] == "Infubinol") | (mouse_study["Drug_Regimen"] == "Ceftamin")]

# Start by getting the last (greatest) timepoint for each mouse
max_time = best_drug_mouse.groupby("Mouse_ID")["Timepoint"].max()

#Merge this group df with the original DataFrame to get the 
#tumor volume at the last timepoint
best_drug_mouse = pd.merge(best_drug_mouse, max_time, how = "outer", on = "Mouse_ID")
best_drug_mouse = best_drug_mouse.rename(columns = {
    "Timepoint_x": "Timepoint",
    "Timepoint_y": "Final_Timepoint"
})
```

```python
#Put treatments into a list for for loop (and later for plot labels)
drug_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_data = []
drug_tumor = []

#Calculate the IQR and quantitatively determine if there are 
#any potential outliers. 
for drug in drug_regimens:
    
    #Locate the rows which contain mice on each drug and get 
    #the tumor volumes
    drug_data = best_drug_mouse.loc[best_drug_mouse["Drug_Regimen"] == drug]
    mouse_ids = list(drug_data["Mouse_ID"].unique())
    
    for mouse in mouse_ids:
        
        final_timepoint_list = list(drug_data.loc[drug_data["Mouse_ID"] == mouse]["Final_Timepoint"])
        final_tumor = drug_data.loc[(drug_data["Mouse_ID"] == mouse) & (drug_data["Timepoint"] == final_timepoint_list[0]), "Tumor_Volume_(mm3)"].iloc[0]
        drug_tumor.append(final_tumor)
    
    #Determine outliers using upper and lower bounds
    quartiles = np.quantile(drug_tumor, q = np.arange(0.25, 1, 0.25))
    print(quartiles)
    
    lowerq = quartiles[0]
    upperq = quartiles[2]
    iqr = upperq - lowerq

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outliers = drug_data.loc[(drug_data["Tumor_Volume_(mm3)"] < lower_bound) | (drug_data["Tumor_Volume_(mm3)"] > upper_bound)]["Tumor_Volume_(mm3)"]
    print(f"{drug}'s potential outliers: {outliers}") 
    
    tumor_data.append(drug_tumor)
    drug_tumor = []
    
###see phone tabs for stackoverflow
```

```python
#Generate a box plot that shows the distrubution of the 
#tumor volume for each treatment group.
fig, ax1 = plt.subplots()
ax1.boxplot(tumor_data, labels = drug_regimens, flierprops = {"markersize": 12, "markerfacecolor": "red"})
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.set_title("Drug Regimen vs. Final Tumor Volume")
plt.show
```

```python
#Generate a scatter plot of mouse weight vs. the average 
#observed tumor volume for the entire Capomulin regimen
capomulin_data = mouse_study.loc[mouse_study["Drug_Regimen"] == "Capomulin"]
capomulin_data = capomulin_data.groupby("Mouse_ID")

capomulin_data_avg_tumor_volume = capomulin_data["Tumor_Volume_(mm3)"].mean()
capomulin_data_weight = capomulin_data["Weight_(g)"].mean()

plt.scatter(capomulin_data_weight, capomulin_data_avg_tumor_volume)

plt.title("Weight vs. Average Tumor Volume of Capomulin Mice")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
#Calculate the correlation coefficient and a linear regression model 
#for mouse weight and average observed tumor volume for 
#the entire Capomulin regimen

plt.scatter(capomulin_data_weight, capomulin_data_avg_tumor_volume)

(slope, intercept, rvalue, pvalue, stderr) = linregress(capomulin_data_weight, capomulin_data_avg_tumor_volume)
regress_values = (capomulin_data_weight * slope) + intercept

line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
plt.plot(capomulin_data_weight, regress_values, "r-")
plt.annotate(line_eq, (5.8,0.8), fontsize=15, color="red")

plt.title("Weight vs. Average Tumor Volume of Capomulin Mice")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

correlation = st.pearsonr(capomulin_data_weight, capomulin_data_avg_tumor_volume)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0], 2)}")
```
